```python
import hvplot.pandas
import pandas as pd
import requests
import cartopy
import geoviews
import pyproj
import holoviews
# Import API key
from api_keys import geoapify_key
```

```python
#load csv
poverty_df= pd.read_csv("../Challenges/School_Neighborhood_Poverty_Estimates_-_Current.csv")

poverty_df.head()
```

```python
#create dataframe with only necessary columns
clean_poverty_df= poverty_df[["NAME","IPR_SE","SCHOOLYEAR","LAT","LON"]]
#pulling a random sample of 5000 rows
random_poverty_df = clean_poverty_df.sample(n=5000, random_state=42)
```

```python
#map plot

#The IPR indicator ranges from 0 to 999. 
#1 Low IPR values indicate a high degree of poverty. 
#A family with income at the poverty threshold has an IPR value of 100. 

map_plot = random_poverty_df.hvplot.points(
    "LON", 
    "LAT", 
    geo = True, 
    size = "IPR_SE",
    scale = .4,
    color = "IPR_SE",
    cmap = "YlOrRd",
    alpha = 0.5,
    tiles = "CartoLight",
    frame_width = 700,
    frame_height = 500
).opts(tools=["wheel_zoom"])


# Display the map
map_plot
```
