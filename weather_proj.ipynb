```python
# Define a dictionary to map severity values to numerical values
severity_mapping = {
    "Light": 1,
    "Severe": 2,
    "Moderate": 3,
    "UNK": 4,
    "Other": 5,
    "Heavy": 6
}

# Add a new column 'n_severity' using the map function
weather_df['n_severity'] = weather_df['Severity'].map(severity_mapping)
```

```python
some_weather_df=weather_df.sample(n=5000, random_state=42)
map_plot = some_weather_df.hvplot.points(
    "LocationLng", 
    "LocationLat", 
    geo = True, 
    size = "n_severity",
    scale = 2,
    color = "Type",
    #cmap = "Plasma",
    alpha = 0.5,
    tiles = "CartoLight",
    frame_width = 700,
    frame_height = 500
).opts(tools=["wheel_zoom"])

map_plot
```

```python
# Filter rows with "Heavy" severity
heavy_weather_df = some_weather_df[some_weather_df['Severity'] == 'Heavy']

# Group by latitude and longitude and reset the index
grouped_heavy_weather_df = heavy_weather_df.groupby(['LocationLat', 'LocationLng']).agg(
    Severity_Count=('Severity', 'size'),
    n_severity=('n_severity', 'mean')
).reset_index()
grouped_heavy_weather_df
```
